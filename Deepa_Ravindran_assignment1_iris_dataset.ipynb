#1.PIE Plot of Classification

```python
#returns a Series having count of unique values present in column named 'Classification'...

df['Classification'].value_counts()
```

```python
#to get the index values of above returned Series and using it as the labels of pie graph...

label=df['Classification'].value_counts().index
label
```

```python
#ploting the Pie graph of 'Classification' column ...

plt.figure(figsize=(6,8))
textprop={'fontsize':15}#to set the font size of labels...
plt.pie(df['Classification'].value_counts(),labels=label,autopct='%1.1f%%',textprops=textprop)
plt.title('IRIS CLASSIFICATION',fontsize=20,c='b')
plt.legend(loc=(1.02,0));#showing legend outside the plot..
```

###From the above plot we can conclude that

####THERE ARE 3 DIFFERNT TYPES OF IRIS FLOWER ARE PRESENT IN THE GIVEN DATASET 
####1.IRIS-SETOSA-50 nos(33.3%)
####2.IRIS-VIRGINICA-50 nos(33.3%)
####3.IRIS-VERSICOLOR-50 nos(33.3%)


#2.Hist plot of SL, SW, PL, PW values...

```python
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.hist(df['SL'],color='g',rwidth=.8)
plt.title('Sepal Length ',fontsize=20,c='b');

plt.subplot(2,2,2)
plt.hist(df['SW'],color='g',rwidth=.8)
plt.title('Sepal Width ',fontsize=20,c='b');

plt.subplot(2,2,3)
plt.hist(df['PL'],color='g',rwidth=.8)
plt.title('Petal Length ',fontsize=20,c='b');

plt.subplot(2,2,4)
plt.hist(df['PW'],color='g',rwidth=.8)
plt.title('Petal Width ',fontsize=20,c='b');

plt.tight_layout()
```

###From the above plot we can conclude the following...

####1.The highest frequency of the sepal length is between 25 and 30 which is between 5.5 and 6
####2.The highest frequency of the sepal Width is between 35 and 40 which is between 3.0 and 3.5
####3.The highest frequency of the petal length is between 35 and 40 which is between 1 and 2
####4.The highest frequency of the petal width is between 40 and 45 which is between 0.0 and 0.5

#3.Scatter plots 

```python
#scatter plot of sl vs sw...

plt.figure(figsize=(10,6))
plt.subplot(1,2,1);
sns.scatterplot(x=df['SL'],y=df['SW'],hue=df['Classification']);
plt.xlabel('SL',fontsize=15,c='b')
plt.ylabel('SW',fontsize=15,c='b')
plt.title('Scatter Plot of SL vs SW',fontsize=20,c='b')


#scatterplot of pl vs pw...

plt.subplot(1,2,2);
sns.scatterplot(x=df['PL'],y=df['PW'],hue=df['Classification']);
plt.title('Scatter plot of PL vs PW',fontsize=20,c='b')
plt.xlabel('PL',fontsize=15,c='b')
plt.ylabel('PW',fontsize=15,c='b')

plt.tight_layout()
```

###From the above plot we can conclude the following...

#a)SL vs SW

####1.Iris-setosa flower has the smallest sepal lengths but the largest sepal widths.

####2.Iris-versicolor flower has lies in the middle of other two calssifications in the terms of sepal lengths ans sepal widths.

####3.Iris-virginica has the largest sepal lengths and sepal widths ranges in the middle.

#b)PL vs PW

####1.Iris-Setosa has the smallest  petal lengths and petal widths among the three flower classifications given.

####2.Iris-Versicolor Species lies in the middle of the other two flower classifications in terms of petal lengths and petal widths.

####3.Iris-Virginica has the largest petal lengths and petal widths among the three flower classifications given.

#5.BOX Plot

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
sns.boxplot(x='Classification',y=df['SL'],data=df)
plt.title('Boxplot of SL vs Classifications',c='b')

plt.subplot(2,2,2)
sns.boxplot(x='Classification',y=df['SW'],data=df)
plt.title('Boxplot of SW vs Classifications',c='b')

plt.subplot(2,2,3)
sns.boxplot(x='Classification',y=df['PL'],data=df)
plt.title('Boxplot of PL vs Classifications',c='b')

plt.subplot(2,2,4)
sns.boxplot(x='Classification',y=df['PW'],data=df);
plt.title('Boxplot of PW vs Classifications',c='b')

plt.tight_layout()
```

#From the above graph, we can conclude that ...
####1.Iris-setosa has the smallest features and less distributed values for SL, PL, and PW. It has some outlier values for PL, and PW.
####2.Iris-veriscolor has average values for SL, SW, PL, PW and outlier values for PL.
####3.Iris-virginica has largest values for SL, PL, PW  and outlier values for SL and SW.
